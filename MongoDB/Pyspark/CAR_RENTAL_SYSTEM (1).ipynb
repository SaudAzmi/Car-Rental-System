```
df= spark.read.csv("/FileStore/tables/CarRentalData_1-1.csv", header=True, inferSchema=True)
```

```
display(df)
```

```
df.createOrReplaceTempView("Car_Rental")
```

```
spark.sql("SELECT * FROM Car_Rental")
```

```
%sql
select* from`Car_Rental`
```

```
%sql
/* Top 3  states that uses the most cars :*/
select state,count(state) as Count from `Car_Rental` group by 1 order by Count desc
limit 3
```

```
%sql
/* Top 3 cities that uses the most cars :*/
select city,count(city) as Count from `Car_Rental` group by 1 order by Count desc
limit 3
```

```

%sql
/*Top states with max owners*/
select state,id,count(state) as Count from `Car_Rental` group by state,id   order by Count desc
```

```
%sql
/*different vehicle:*/
select vehicletype,count(vehicletype) as Count from `Car_Rental` group by vehicletype   order by Count desc
```

```
%sql
/* Top 5 vehicle makers who helped making most avg. daily rate*/
select vehiclemake ,avg(ratedaily) as avgrate from `Car_Rental` group by vehiclemake   order by avgrate desc
limit 5
```

```
%sql
/*different vehicle:*/
select fuelType,count(vehicletype) as Count from `Car_Rental` group by fuelType   order by Count desc
```

```
%sql
/* Top 5 vehicle makers who helped making most avg. daily rate*/
select fuelType,avg(renterTripsTaken) as avgtrip from `Car_Rental` group by fuelType   order by avgtrip desc
limit 5
```
