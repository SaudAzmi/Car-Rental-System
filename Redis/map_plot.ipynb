```python
import pandas as pd
import plotly.express as px
import pymongo
```

```python
cars_df = pd.read_csv('cars.csv')
```

```python
customers_df = pd.read_csv('customers.csv')
drivers_df = pd.read_csv('drivers.csv')
drop_df = pd.read_csv('drop.csv')
```

```python
cars_df.head()
```

```python
cars_df = cars_df.drop(columns=['Unnamed: 0', '_id', 'type', 'available'])
```

```python
cars_df = cars_df.rename(columns={'location_name':'_id'})
```

```python
cars_df.head()
```

```python
customers_df = customers_df.drop(columns=['Unnamed: 0', 'searching'])
```

```python
drivers_df = drivers_df.drop(columns=['Unnamed: 0', 'available', 'rating'])
```

```python
drop_df = drop_df.drop(columns=['Unnamed: 0', '_id'])
```

```python
drop_df = drop_df.rename(columns={'location_name':'_id'})
```

```python
drop_df.head()
```

```python
drivers_df.shape
```

```python
full_df = pd.concat([cars_df, customers_df, drivers_df, drop_df])
```

```python
full_df.head()
```

```python

fig = px.scatter_mapbox(
    full_df.loc[full_df['_id'].isin([6133741255, 'Metz-Tessy', 'Vinzier', 6227049953])],  # Our DataFrame
    lat = "lat",
    lon = "lon",
    center = {"lat": 45.9363312, "lon": 6.1080329},  # where map will be centered
    width = 600,  # Width of map
    height = 600,  # Height of map
    color = ['Vehicle', 'customer', 'Driver', 'Destination'],
    size = [5,5,5,5],
    hover_data = ["_id"]  # what to display when hovering mouse over coordinate
)

fig.update_layout(mapbox_style="open-street-map") # adding beautiful street layout to map

fig.show()
```
